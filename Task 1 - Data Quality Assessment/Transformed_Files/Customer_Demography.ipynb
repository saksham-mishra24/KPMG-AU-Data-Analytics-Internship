```python
Customer_Demography = Customer_Demography.drop('default', axis='columns')
```

```python
Customer_Demography['tenure'].fillna(value=11.0,inplace= True)
```

```python
# Drop columns with fewer non-null values than the threshold (50% of the number of rows)
drop_threshold = Customer_Demography.shape[0] * 0.5
Customer_Demography.dropna(thresh=drop_threshold, axis='columns', inplace=True)
```

```python
# check the frequency of different job titles
Customer_Demography['job_title'].value_counts()
```

```python
Customer_Demography['job_title'] = Customer_Demography['job_title'].fillna(method='bfill')
```

```python
# replace all the misspelt or differently formatted data
Customer_Demography['gender'].replace({'M' : 'Male', 'F' : 'Female', 'Femal' : 'Female'}, inplace=True)
```

```python
# make sure all dates are in the correct format
Customer_Demography['DOB'] = Customer_Demography['DOB'].dt.date
```

```python
# fill the missing datapoint using the forward filling method
Customer_Demography['job_industry_category']= Customer_Demography['job_industry_category'].fillna(method='ffill')
Customer_Demography['job_industry_category']= Customer_Demography['job_industry_category'].fillna(method='ffill')

```

```python
# make sure that the column names are in the correct format
Customer_Demography.columns = map(str.lower, Customer_Demography.columns)
Customer_Demography.columns = map(str.strip, Customer_Demography.columns)
```
